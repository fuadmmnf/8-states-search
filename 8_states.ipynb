```python
class PriorityQueue(object): 
    def __init__(self): 
        self.queue = [] 
  

    # for checking if the queue is empty 
    def isEmpty(self): 
        return len(self.queue) == [] 
  
    # for inserting an element in the queue 
    def append(self, data): 
        self.queue.append(data) 
  
    # for popping an element based on Priority 
    def pop(self): 
        global node_cost
        try: 
            min = 0
            for i in range(len(self.queue)): 
                current_min_cost, current_cost = 0, 0
                if tuple(self.queue[i]) in node_cost:
                    current_cost = node_cost[tuple(self.queue[i])]
                if tuple(self.queue[min]) in node_cost:
                    current_min_cost = node_cost[tuple(self.queue[min])]
                
                if current_cost + greedyHeuristic(self.queue[i]) < current_min_cost + greedyHeuristic(self.queue[min]): 
                    min = i 
            item = self.queue[min] 
            del self.queue[min] 
            return item 
        except IndexError: 
            print() 
            exit() 
```

```python
def greedyHeuristic(board_state):
    global goal_state
    manhattan_distance = 0
    for i in range(9):
        goal_index = [i/3, i%3]
        board_index = [board_state.index(i)/3, board_state.index(i)%3]
        manhattan_distance += abs(goal_index[0] - board_index[0]) + abs(goal_index[1] - board_index[1])
#         manhattan_distance += (i-board_state[i])**2
    return manhattan_distance
        
 
```

```python
def print_solution(path_list):
    if path_list is None:
        print("No Solution found")
        return 
    
    prev_board = None
    for path in path_list:
        board = path.copy()
        
        if prev_board == None:
            move = "Initial Stage"
        else:
            present_index = board.index(0)
            previous_index = prev_board.index(0)
            
            if present_index - previous_index == 3:
                move = "Moved Down"
            if present_index - previous_index == -3:
                move = "Moved Up"
            if present_index - previous_index == 1:
                move = "Moved Right"
            if present_index - previous_index == -1:
                move = "Moved Left"
            
        board_view ="._______.\n"+f"| {board[0]} {board[1]} {board[2]} | {move}.\n"+f"| {board[3]} {board[4]} {board[5]} | \n"+f"| {board[6]} {board[7]} {board[8]} | \n"+f"---------"
        print(board_view, end="\n\n")
        
        prev_board = board
```

```python
def get_swapped_state(present_state, direction):
#     print('present ',present_state)
    temp_state = present_state
    blank_index = temp_state.index(0)

    if direction == "up":
        temp_state[blank_index-3], temp_state[blank_index] = temp_state[blank_index], temp_state[blank_index-3] 
        
    elif direction == "down":
        temp_state[blank_index+3], temp_state[blank_index] = temp_state[blank_index], temp_state[blank_index+3] 

    elif direction == "left":
        temp_state[blank_index-1], temp_state[blank_index] = temp_state[blank_index], temp_state[blank_index-1] 

    elif direction == "right":
        temp_state[blank_index+1], temp_state[blank_index] = temp_state[blank_index], temp_state[blank_index+1] 
        
#     print('after ',temp_state)        
    return temp_state
```
